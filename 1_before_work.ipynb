# 创建各时期的数据文件夹，用于存放数据

```python
def creat_multi_file(path,name_dir):
    import os 
    os.chdir(path)
    isExists = os.path.exists(name_dir)  #判断要创建的文件夹是否存在，不存在则创建，存在则进行下一步文件的写入
    if not isExists:
        os.mkdir(name_dir)
        print(name_dir + " 文件夹创建成功")
        print("______________________________\n")
    else:
        print(name_dir + "文件夹已经存在，正在为你进行下一步......")
        print("______________________________\n")
        
    #写入文件
    ls_1 = list(range(2004,2020))  #数据的起始年份和终止年份
    ls_2 = list(range(1,12)) #1 - 11月
    for i in ls_1:
        for j in ls_2[::2]:
            name = str(i)+"_"+str(j)+"-"+str(j+1) #设置文件名格式
                
            os.chdir(path + '/' + name_dir)
            f = open("{}.xls".format(name),"w")  
            f.close()
            print("{}.xls".format(name)," 文件创建成功")
```

```python
creat_multi_file("C:/Users/86177/Desktop/新建文件夹/","中国安全生产事故数据")
```

```python
import pathlib
import pandas as pd
```

```python
path_dir = 'C:\\Users\\86177\\Desktop\\新建文件夹\\中国安全生产事故数据'
```

```python
path_root = pathlib.Path(path_dir)
path_root#生成一个路径对象
```

```python
for item in path_root.iterdir():#对里面所有的文件进行迭代
    print(item)
#查看包含内容
type(item)
#路径是windowpath对象，不是直接可以使用的绝对路径
```

```python
#将所有的路径数据存放到列表中
path_list = []
for item in path_root.iterdir():#对里面所有的文件进行迭代
    path_list.append(item)
len(path_list)
```

```python
all_file_paths = [str(path) for path in path_list]
all_file_paths[:3]
#将所有的windowpath转化为绝对路径
```

```python
p = 'C:\\Users\\86177\\Desktop\\新建文件夹\\中国安全生产事故数据\\2003_11-2004_2.xls'
t = p.split("\\")[-1].split("-")[0]
#type(p.split("\\")[-1][:7])  是字符串类型
t
```

```python
ls_name = []
ls_date = []
ls_value = []

for path in all_file_paths:
    data = pd.read_excel(path)
    ls_name.append(data["省市"].values[:-2])
    ls_value.append(data["事故/起"].values[:-2])
    ls_date.append([path.split("\\")[-1].split("-")[0].replace("_","/") + "/01"]*30)
```

```python
name_data = []
date_data = []
value_data = []
    
for num in range(len(ls_date)):
    for i,j,k in zip(ls_name[num],ls_value[num],ls_date[num]):
        name_data.append(i)
        value_data.append(j)
        date_data.append(k)
        
df = pd.DataFrame({"name": name_data,"value":value_data,"date":date_data})
df
```
